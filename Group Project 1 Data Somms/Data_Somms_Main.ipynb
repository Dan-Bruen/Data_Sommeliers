```python
# Dependencies
import pandas as pd
```

```python
# Storing our Data World Wine filepath in a variable
data_world_wines = "DataWorldWines.csv"
```

```python
# Preview the Data World Wine dataframe
data_world_wines_df = pd.read_csv(data_world_wines, encoding="ISO-8859-1")
data_world_wines_df.head()
```

```python
# Summary Data
## By Vintage - number of wines per Vintage
data_world_wines_df["Vintage"].value_counts()
```

```python
## Average Points per Vintage
data_world_wines_df.groupby("Vintage").mean()
```

```python
# Avg Points by Variety/Grape
data_world_wines_df.groupby("Variety").mean()
```
